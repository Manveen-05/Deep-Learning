```python
# Select quote column from dataframe
quotes = df['quote']

# Display first 5 quotes
quotes.head()
```

```python
quotes = quotes.str.lower()
```

```python
# Import string module
import string

# Create translation table
# Removes all punctuation marks
translator = str.maketrans(
    '',
    '',
    string.punctuation
)

# Apply punctuation removal to all quotes
quotes = quotes.apply(
    lambda x: x.translate(translator)
)
```

```python
quotes.head()
```

```python
# Import Tokenizer
# Used to convert text into numerical sequences
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# Set maximum vocabulary size
vocab_size = 10000

# Create tokenizer
# Keeps top 10,000 most frequent words
tokinizer = Tokenizer(
    num_words=vocab_size
)

# Build vocabulary from quotes
tokinizer.fit_on_texts(quotes)
```

```python
# Get dictionary of words and their index values
word_index = tokinizer.word_index

# Print total unique words
print(len(word_index))

# Display first 10 words with indices
list(word_index.items())[:10]
```

```python
# Convert text quotes into numerical sequences
# Each word is replaced with its corresponding token index
sequence = tokinizer.texts_to_sequences(quotes)
```

```python
# Create empty lists for input and output sequences
X = []
y = []

# Loop through each tokenized sentence
for seq in sequence:

    # Create sequence pairs
    for i in range(1, len(seq)):

        # Input sequence
        # Example: [12, 45]
        input_seq = seq[:i]

        # Output word
        # Example: 67
        output_seq = seq[i]

        # Store input and output
        X.append(input_seq)
        y.append(output_seq)
```

```python
# Find maximum sequence length
# Used later for padding sequences
max_len = max(len(x) for x in X)

# Display maximum length
print(max_len)
```

# Padding

```python
# Import pad_sequences
# Used to make all sequences the same length
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Pad input sequences
# 'pre' adds zeros at the beginning
X_padded = pad_sequences(
    X,
    maxlen=max_len,
    padding='pre'
)
```

```python
# Import Sequential model
# Used to build neural network layer-by-layer
from tensorflow.keras.models import Sequential

# Import deep learning layers
from tensorflow.keras.layers import (
    Embedding,   # Converts words into dense vectors
    SimpleRNN,   # Basic Recurrent Neural Network layer
    LSTM,        # Long Short-Term Memory layer
    Dense        # Fully connected output layer
)
```

```python
# Size of embedding vectors
# Each word will be represented using 50 features
embedding_dim = 50

# Number of neurons in RNN/LSTM layer
# Higher value increases learning capacity
rnn_units = 128
```

# RNN Model

```python
# Create Sequential RNN model
# Layers will be added one after another
rnn_model = Sequential()
```

```python
# Add Embedding layer
# Converts word indices into dense vector representations
rnn_model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        input_length=max_len
    )
)

# Add Simple RNN layer
# Learns sequential patterns from text
rnn_model.add(
    SimpleRNN(
        units=rnn_units
    )
)

# Add Output layer
# Softmax used for multi-class word prediction
rnn_model.add(
    Dense(
        units=vocab_size,
        activation='softmax'
    )
)
```

```python
# Compile RNN model
rnn_model.compile(

    # Adam optimizer for efficient training
    optimizer='adam',

    # Loss function for multi-class classification
    loss='categorical_crossentropy',

    # Evaluation metric
    metrics=['accuracy']
)
```

```python
# Create Sequential LSTM model
lstm_model = Sequential()

# Add Embedding layer
# Converts word indices into dense vectors
lstm_model.add(
    Embedding(
        input_dim=vocab_size,
        output_dim=embedding_dim,
        input_length=max_len
    )
)

# Add LSTM layer
# Learns long-term sequential dependencies
lstm_model.add(
    LSTM(
        units=rnn_units
    )
)

# Add Output layer
# Softmax activation for next-word prediction
lstm_model.add(
    Dense(
        units=vocab_size,
        activation='softmax'
    )
)
```

```python
# Create empty dictionary
# Used to convert index back to words
index_to_word = {}

# Loop through word-index pairs
for word, index in word_index.items():

    # Store reverse mapping
    # Example: 5 -> "hello"
    index_to_word[index] = word
```

```python

from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# Function to predict next word
def predictor(model, tokenizer, text, max_len):

    # Convert input text to lowercase
    text = text.lower()

    # Convert text into token sequence
    seq = tokenizer.texts_to_sequences([text])[0]

    # Pad sequence to match training length
    seq = pad_sequences(
        [seq],
        maxlen=max_len,
        padding='pre'
    )

    # Predict probabilities for next word
    pred = model.predict(
        seq,
        verbose=0
    )

    # Get index with highest probability
    pred_index = np.argmax(pred)

    # Convert predicted index back to word
    return index_to_word[pred_index]
```

```python
# Input sentence
seed_text = "what are you"

# Predict next word using trained LSTM model
next_word = predictor(
    lstm_model,
    tokinizer,
    seed_text,
    max_len
)

# Display predicted word
print(next_word)
```

```python
# Function to generate text sequence
def generate_text(model, tokenizer, seed_text, max_len, n_words):

    # Generate words one by one
    for _ in range(n_words):

        # Predict next word
        next_word = predictor(
            model,
            tokenizer,
            seed_text,
            max_len
        )

        # Stop if no valid word is predicted
        if next_word == "":
            break

        # Add predicted word to sentence
        seed_text += " " + next_word

    # Return final generated text
    return seed_text
```

```python
# Starting input text
seed = "are you a "

# Generate next 10 words using LSTM model
generated_text = generate_text(
    lstm_model,
    tokinizer,
    seed,
    max_len,
    10
)

# Display generated sentence
print(generated_text)
```
